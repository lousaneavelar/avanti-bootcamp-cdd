```python
import pandas as pd

# Carregar o conjunto de dados
df = pd.read_csv('../data/raw/penguins2.csv')

# Visualizar as primeiras linhas do DataFrame
df.head()
```

```python
# Descrição das variáveis
descricao = df.describe(include='all')

# Verificar a quantidade de valores faltantes
valores_faltantes = df.isnull().sum()

# Combinar a descrição e os valores faltantes em um único DataFrame
descricao.loc['Valores Faltantes'] = valores_faltantes


# Visualizar a descrição completa
print("\nDescrição completa das variáveis e valores faltantes:")
print(descricao)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Gráfico de contagem das espécies de pinguins
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='species')
plt.title('Contagem de Espécies de Pinguins')
plt.show()
```

```python
# Análise da relação entre comprimento do bico e massa
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='bill_length_mm', y='body_mass_g', hue='species')
plt.title('Relação entre Comprimento do Bico e Massa Corporal')
plt.xlabel('Comprimento do Bico (mm)')
plt.ylabel('Massa Corporal (g)')
plt.legend(title='Espécie')
plt.show()
```
